```python
url = "https://www.nseindia.com/products/content/equities/indices/historical_index_data.htm"
page = requests.get(url)
stocks = []
soup = BeautifulSoup(page.text,'html.parser')
mylist = soup.find_all(class_="viewdata-content")[0]
eachlist = mylist.find_all('option')
for alllist in eachlist:
    stocks.append(alllist['value'])
```

```python
stocks.pop(0)
```

```python
stocks
```

```python
def calculate_moving_average(mera_column):
    df = pd.DataFrame(mera_column)
    print(df.rolling(window=28).mean())


for each_stock in stocks:
    each_stock_data = np.get_history(symbol=each_stock, start=date(2015,1,1), end=date(2015,12,31),index=True)
    calculate_moving_average(each_stock_data['Close'])
```
